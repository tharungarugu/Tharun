Preparing the training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=1)
```

Displaying the content of the X_train

```python
logis=LogisticRegression()
```

```python
logis.fit(X_train,Y_train)
```

```python
predicted=logis.predict(X_train)
```

Finding the accuracy of training dataset

```python
predicted_X_train=accuracy_score(predicted,Y_train)
```

```python
predicted_X_train
```

```python
X_test_predicted=logis.predict(X_test)
```

```python
predicted_X_test_accuracy=accuracy_score(X_test_predicted,Y_test)
```
